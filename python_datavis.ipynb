```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
import numpy as np
```

```python
df = pd.read_csv('Churn Modeling.csv')
```

```python
#Dropping columns that won't help in our analysis:
clean_df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
clean_df
```

```python
pxfig = px.scatter(clean_df, x='EstimatedSalary', y='Balance', color='Exited')
pxfig.show()
```

```python
#Saving the plot
pio.write_image(pxfig, "PxScatter2.jpg", width=1200, height=400, scale=1)
```

```python
#Seaborn Corrmat
corrmat = clean_df.corr()
fig, ax = plt.subplots(figsize = (20, 10))
sns.heatmap(corrmat, vmin=corrmat.values.min(), vmax=1, square=True, annot=True, linewidths=0.1, cmap="YlGnBu", fmt='0.2f')
```

```python
#Seaborn Pairplot

pairplot_cols = ['Exited', 'Age', 'Balance', 'Tenure']
sns.pairplot(clean_df[pairplot_cols], height=3)
plt.show()
```

```python
g = sns.catplot(data=clean_df, x='Age', y='Tenure', hue='Exited', jitter=True)
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(12,10)
plt.show()
```
